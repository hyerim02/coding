# 3. 개미 전사

개미: 식량 부족해서 메뚜기 식량 창고 공격  
메뚜기: 일직선으로 연결되어 있는 여러 개의 식량창고를 갖고 있고 인접한 식량창고가 공격받으면 바로 알 수 있음

최소한 한 칸 이상 떨어진 식량창고를 약탈해야 함

- 입력 조건  
첫째 줄에 식량창고 개수 N이 주어짐(3이상 100이하)  
둘째 줄에 식량의 개수 K가 주어짐(0이상1000이하) 

- 출력 조건  
식량의 최댓값

```python
n=int(input())
```

```python
array=list(map(int,input().split()))
```

```python
d=[0]*100
d[0]=array[0]
d[1]=max(array[0],array[1])
for i in range(2,n):
    d[i]=max(d[i-1],d[i-2]+array[i])
```

```python
d[n-1]
```

# 4. 바닥 공사

바닥을 채우는 경우의 수를 구하는 프로그램을 작성

- 입력 조건  
N이 주어짐(1이상1000이하)

- 출력 조건  
경우의 수를 796796으로 나눈 나머지를 출력

```python
n=int(input())
```

```python
d=[0]*1001
d[1]=1
d[2]=3
for i in range(3,n+1):
    d[i]=(d[i-1]+2*d[i-2])%796796
```

```python
d[n]
```

# 5. 효율적인 화폐 구성

N가지 종류의 화폐가 있고 최소한 개수를 사용하여 M이 되도록 함

각 화폐는 몇 개라도 사용할 수 있음  
구성은 같은데 순서만 다른 것은 같은 경우로

- 입력 조건  
첫째 줄에 N,M이 주어짐(1이상100이하, 1이상 만이하)  
만 이하의 화폐의 가치가 주어짐

- 출력 조건  
최소한의 화폐 개수  
불가능한 것은 -1

```python
n,m=map(int,input().split())
```

```python
array=[]
for i in range(n):
    array.append(int(input()))
```

```python
d = [10001] * (m + 1)
d[0] = 0
for i in range(n):
    for j in range(array[i], m + 1):
        if d[j - array[i]] != 10001:
            d[j] = min(d[j], d[j - array[i]] + 1)


if d[m] == 10001: 
    print(-1)
else:
    print(d[m])
```
