## 예제 3-1 거스름돈

당신은 음식점의 계산을 도와주는 점원이다. 카운터에는 거스름돈으로 사용할 500원,100원,50원,10원 동전이 무한히 존재한다고 가정  
손님에게 거슬러 줘야 할 돈이 N원일 때 거슬러 줘야 할 동전의 최소의 개수를 구해라.  
단,거슬러 줘야 할 돈 N은 항상 10의 배수

- N=1260

```python
n=1260
count=0
coin=[500,100,50,10]
for i in coin:
    count+=n//i
    n%=i
print(count)    
```

## 큰 수의 법칙

다양한 수로 이루어진 배열이 있을 때 주어진 수들을 M번 더하여 가장 큰 수를 만드는 법칙  
단, 배열의 특정한 인덱스에 해당하는 수가 연속해서 K번을 초과하여 더해질 수 없는 것이 특징  
같은 수여도 인덱스가 다르면 다른걸로 인정

- 입력 조건  
첫째 줄에 N(2<=N<=1000),M(1<=M<=10000),K(1<=K<=10000)의 자연수가 주어지며, 각 자연수는 공백으로 구분  
둘째 줄에 N개의 자연수가 주어짐, 각 자연수는 공백으로 구분, 각각의 자연수는 1 이상 10000이하의 수  
K<=M

- 출력 조건  
합을 출력

```python
n,m,k=map(int,input().split())
```

```python
data=list(map(int,input().split()))
```

```python
data.sort()
first=data[n-1]
second=data[n-2]

count=int(m/(k+1))*k
count+=m%(k+1)

result=0
result+= (count)*first
result+=(m-count)*second

print(result)
```

## 숫자 카드 게임

여러 개의 숫자 카드 중에서 가장 높은 숫자가 쓰인 카드 한 장을 뽑는 게임

1. 숫자가 쓰인 카드들이 N $\times$ M 형태로 놓여 있음  
2. 먼저 뽑고자 하는 카드가 포함되어 있는 행을 선택  
3. 그다음 선택된 행에 포함된 카드들 중 가장 숫자가 낮는 카드를 뽑아야 함  
4. 처음 카드를 골라낼 행을 선택할 때, 이후에 해당 행에서 가장 숫자가 낮은 카드를 뽑을 것을 고려하여 최종적으로 가장 높은 숫자의 카드를 뽑을 수 있도록 전략을 세워야 함

예를 들어서  
3 1 2  
4 1 4  
2 2 2  
카드가 이렇게 놓여 있으면 첫 번째 혹은 두 번째 행은 최종적으로 1을 뽑게 되고 세 번째 행은 2를 뽑게 된다 

- 입력 조건  
첫째 줄에 숫자 카드들이 놓인 행의 개수 N과 열의 개수 M이 공백을 기준으로 하여 가각 자연수로 주어짐(1<=N,M<=100)  
둘째 줄부터 N개의 줄에 걸쳐 각 카드에 적힌 숫자가 주어지고 각 숫자는 1 이상 10000이하인 자연수

- 출력 조건  
카드 숫자 출력

```python
n,m=map(int,input().split())
```

```python
result=0
for i in range(n):
    data=list(map(int,input().split()))
    min_v=min(data)
    result=max(result,min_v)
print(result)    
```

# 1이 될 때까지

어떠한 수 N이 1이 될 때까지  
1. -1을 하기  
2. k로 나누기  
둘 중 하나를 반복적으로 선택하여 수행  
단. 2번은 나누어떨어질 때만

N과 k가 주어질 때 1 또는 2를 수행하는 최소 횟수를 구해라 

- 입력 조건  
첫째 줄에 N(2<=N<=100000)과 k(2<=k<=100000)가 공백으로 구분되며, 각각 자연수로 주어짐   
N >= k

- 출력 조건  
횟수 출력

```python
n,k=map(int,input().split())
```

```python
result=0
while n>=k:
    while n % k !=0:
        n-=1
        result+=1
    n//=k
    result+=1
while n>1:
    n-=1
    result+=1
print(result)    
```
