# 1. 다이나믹 프로그래밍

컴퓨터는 연산 속도의 한계가 있거, 메모리 공간을 사용할 수 있는 데이터의 개수도 한정적이라는 점이 많은 제약을 발생  
==>속도와 메모리 공간을 최대한 활용할 수 있는 효율적인 알고리즘을 작성

메모리 공간을 약간 더 사용하면 연산 속도를 비약적으로 증가시킬 수 있는 방법

- 피보나치 수열  
두 항의 합을 현재의 항으로 설정하는 특징이 있는 수열

$a_n=a_{n-1}+a_{n-2},a_1=1,a_2=1$

이런 수열을 리스트 형태로 표현

- 재귀함수

```python
def fibo1(x):
    if x==1 or x==2:
        return 1
    return fibo1(x-1)+fibo1(x-2)
```

```python
fibo1(35)
```

이렇게 하면 n이 커질수록 수행 시간이 기하급수적으로 커짐

같은게 3번 반복

- **메모이제이션**  
한 번 구한 결과를 메모리 공간에 저장해서 다시 호출하면 그대로 가져옴  
값을 저장하는 방법이므로 캐싱이라고도 함

```python
d=[0]*36
def fibo2(x):
    if x==1 or x==2:
        return 1
    if d[x]!=0:
        return d[x]
    d[x]=fibo2(x-1)+fibo2(x-2)
    return d[x]
```

```python
fibo2(35)
```

큰 문제를 해결하기 위해 작은 문제를 호출 : 탑다운 방식,하향식  
작은 문제부터 차근차근 답을 도출: 보텀업 방식,상향식

```python
d=[0]*36
d[1]=1
d[2]=1
n=35
for i in range(3,n+1):
    d[i]=d[i-1]+d[i-2]
```

```python
d[35]
```

다이나믹 프로그래밍은 전형적인 보텀업 방식  
보텀업에서 사용되는 리스트는 DP 테이블이라고 부름  
메모제이션은 탑다운 방식  

# 2. 1로 만들기

1. 5로 나누어떨어지면 5로 나눔  
2. 3으로 나누어떨어지면 3으로 나눔  
3. 2로 나누어떨어지면 2로 나눔  
4. 1을 빼기  
연산을 적당히 사용해서 1을 만들려고 함

- 입력 조건  
첫째 줄에 정수 N이 주어짐(1이상 3만이하)

- 출력 조건  
연산을 하는 횟수의 최솟값

```python
x=int(input())
```

```python
d=[0]*30001
for i in range(2,x+1):
    d[i]=d[i-1]+1
    if i%2==0:
        d[i]=min(d[i],d[i//2]+1)
    if i%3==0:
        d[i]=min(d[i],d[i//3]+1)
    if i%5==0:
        d[i]=min(d[i],d[i//5]+1)
        
```

```python
d[x]
```
